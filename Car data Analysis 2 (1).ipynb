```python
import matplotlib.pyplot as plt
```

```python
# Group by 'make' and calculate the mean of 'passengers' for each group
mean_passengers_by_make = new_noduplicate.groupby('Make')['Passengers'].mean()

# Plotting a bar chart
plt.bar(mean_passengers_by_make.index, mean_passengers_by_make.values, color='skyblue')
plt.xlabel('Car Make')
plt.ylabel('Mean Passengers')
plt.title('Mean Passengers by Car Make')
plt.show()
```

```python
# Group by 'year' and count occurrences
year_counts =new_noduplicate.groupby('Year').size().reset_index(name='count')

# Display the counts
print("Year Counts:")
print(year_counts)
```

```python
# Group by 'year' and sum the passengers
year_passenger_sum =new_noduplicate.groupby('Year')['Passengers'].sum().reset_index(name='total_passengers')

# Display the result
print("Year and Total Passengers:")
print(year_passenger_sum)

# Plotting a bar chart
plt.bar(year_passenger_sum['Year'], year_passenger_sum['total_passengers'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.title('Total Passengers per Year')
plt.show()
```
